```python
import pandas

income_raw = pandas.read_excel('earnings-residence-borough.xlsx')

income_raw
```

```python
dropconf = [col for col in income_raw.columns if "conf %" in str(income_raw.iloc[0][col])]
income1 = income_raw.drop(columns = dropconf)

income_cleaned = income1.iloc[:,[1,11,12,13,14,15]]
income_cleaned= income_cleaned.dropna(axis=1, how='all')
income_cleaned
```
