```python
import pandas

wellbeing_raw = pandas.read_excel('personal-well-being-borough.xlsx')

wellbeing_raw
```

```python

yearsremove = ['2016/17', '2017/18', '2018/19']

columnsremove = [col for col in wellbeing_raw.columns if wellbeing_raw.iloc[0][col] in yearsremove]


wellbeing1 = wellbeing_raw.drop(columns=columnsremove)

wellbeing1= wellbeing1.dropna(axis=1, how='all')

wellbeing1
```

```python

rename = {'Life Satisfaction':'2011/12 Life Satisfaction',
    'Unnamed: 3': '2012/13 Life Satisfaction',
    'Unnamed: 4': '2013/14 Life Satisfaction',
    'Unnamed: 5': '2014/15 Life Satisfaction',
    'Unnamed: 6': '2015/16 Life Satisfaction',}

rename = {'Worthwhile':'2011/12 Worthwhile',
    'Unnamed: 12': '2012/13 Worthwhile',
    'Unnamed: 13': '2013/14 Worthwhile',
    'Unnamed: 14': '2014/15 Worthwhile',
    'Unnamed: 15': '2015/16 Worthwhile',}

rename = {'Happiness':'2011/12 Happiness',
    'Unnamed: 21': '2012/13 Happiness',
    'Unnamed: 22': '2013/14 Happiness',
    'Unnamed: 23': '2014/15 Happiness',
    'Unnamed: 24': '2015/16 Happiness',}

rename = {'Anxiety':'2011/12 Anxiety',
    'Unnamed: 30': '2012/13 Anxiety',
    'Unnamed: 31': '2013/14 Anxiety',
    'Unnamed: 32': '2014/15 Anxiety',
    'Unnamed: 33': '2015/16 Anxiety',}

wellbeing1 = wellbeing1.rename(columns=rename)

wellbeing1
```

```python
for year in '2011/12', '2012/13', '2013/14', '2014/15', '2015/16':
    wellbeing1[f'{year} Life Satisfaction'] = pandas.to_numeric(wellbeing1[f'{year} Life Satisfaction'], errors='coerce')
    wellbeing1[f'{year} Worthwhile'] = pandas.to_numeric(wellbeing1[f'{year} Worthwhile'], errors='coerce')
    wellbeing1[f'{year} Happiness'] = pandas.to_numeric(wellbeing1[f'{year} Happiness'], errors='coerce')
    wellbeing1[f'{year} Anxiety'] = pandas.to_numeric(wellbeing1[f'{year} Anxiety'], errors='coerce')

    wellbeing1[f'{year} Overall Wellbeing'] = wellbeing1[f'{year} Life Satisfaction'] + wellbeing1[f'{year} Worthwhile'] + \
                                              wellbeing1[f'{year} Happiness'] - wellbeing1[f'{year} Anxiety']

wellbeing1

```

```python

wellbeinglevel_cleaned_df = wellbeing1.iloc[:,[0,1,-5,-4,-3,-2,-1]]

wellbeinglevel_cleaned_df
```
